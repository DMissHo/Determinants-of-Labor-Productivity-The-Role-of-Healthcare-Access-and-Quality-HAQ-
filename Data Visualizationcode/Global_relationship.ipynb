```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the plot
plt.figure(figsize=(12, 8))

# Scatterplot with region-based color groups
regions = df["Region"].unique()

for region in regions:
    subset = df[df["Region"] == region]
    plt.scatter(
        subset["Healthcare Access and Quality Index (HAQ)"],
        subset["Labor Productivity"],
        label=region,
        s=40, alpha=0.7
    )

# Labels and title
plt.title("Relationship Between Healthcare Access & Quality and Labor Productivity", fontsize=16)
plt.xlabel("Healthcare Access and Quality Index (HAQ)", fontsize=14)
plt.ylabel("Productivity (Output per Hour Worked)", fontsize=14)

# Legend
plt.legend(title="Region", fontsize=10)

# Clean style
plt.grid(alpha=0.3)
plt.savefig("health_productivity_plot.png", dpi=300, bbox_inches="tight")

# Show plot
plt.show()
```
