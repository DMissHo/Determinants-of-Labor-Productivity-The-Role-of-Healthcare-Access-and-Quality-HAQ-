```python
df = pd.read_excel("C:\\Users\\user\\Downloads\\Mergedhaq.xlsx")
```

```python
df.head()
```

```python
# Required years
required_years = {2000, 2010, 2015, 2019}

# Remove rows where HAQ or Productivity is missing
df_clean = df.dropna(subset=[
    "Healthcare Access and Quality Index (HAQ)",
    "Productivity: output per hour worked"
])

# Identify countries that have all required years WITH complete values
valid_countries = (
    df_clean.groupby("Country")["Year"]
            .agg(lambda x: required_years.issubset(set(x)))
)

# Filter to keep only the valid countries
df_filtered = df_clean[df_clean["Country"].isin(valid_countries[valid_countries].index)]

print(df_filtered)
```

```python
df_filtered['Country'].nunique()
```

```python
df_filtered.to_excel('cleanMerged.xlsx', index=False)
```

```python
# --- Clean numeric columns ---
df_filtered["Healthcare Access and Quality Index (HAQ)"] = pd.to_numeric(
    df["Healthcare Access and Quality Index (HAQ)"], errors="coerce"
)
df_filtered["Productivity: output per hour worked"] = pd.to_numeric(
    df["Productivity: output per hour worked"], errors="coerce"
)

# --- Group by Region and Year ---
region_year = (
    df_filtered.groupby(["Region", "Year"])
      [["Healthcare Access and Quality Index (HAQ)", "Productivity: output per hour worked"]]
      .mean()
      .reset_index()
)

# --- Plot for each region ---
regions = region_year["Region"].unique()

for region in regions:
    data = region_year[region_year["Region"] == region]

    plt.figure(figsize=(10, 6))

    # Plot HAQ
    plt.plot(
        data["Year"],
        data["Healthcare Access and Quality Index (HAQ)"],
        marker="o",
        label="Average HAQ"
    )
    
    # Plot Productivity
    plt.plot(
        data["Year"],
        data["Productivity: output per hour worked"],
        marker="o",
        label="Average Productivity"
    )

    plt.title(f"Average HAQ and Productivity Over Time in {region}, 2000-2019")
    plt.xlabel("Year")
    plt.ylabel("Average Value")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.xticks(ticks=[2000, 2005, 2010, 2015, 2020])
    plt.show()
```

```python
import math
# Make sure numeric columns are clean
df_filtered["Healthcare Access and Quality Index (HAQ)"] = pd.to_numeric(
    df_filtered["Healthcare Access and Quality Index (HAQ)"], errors="coerce"
)
df_filtered["Productivity: output per hour worked"] = pd.to_numeric(
    df_filtered["Productivity: output per hour worked"], errors="coerce"
)

# Average per region per year
region_year = (
    df_filtered.groupby(["Region", "Year"])
      [["Healthcare Access and Quality Index (HAQ)",
        "Productivity: output per hour worked"]]
      .mean()
      .reset_index()
)

regions = region_year["Region"].unique()
n = len(regions)

# --- Create subplots in 1 row, multiple columns ---
fig, axes = plt.subplots(1, n, figsize=(3*n, 2), sharey=False)

# If only 1 region, axes is not a list, so wrap it
if n == 1:
    axes = [axes]

for ax, region in zip(axes, regions):
    subset = region_year[region_year["Region"] == region]

    ax.plot(subset["Year"], subset["Healthcare Access and Quality Index (HAQ)"], marker="o", label="HAQ")
    ax.plot(subset["Year"], subset["Productivity: output per hour worked"], marker="o", label="Productivity")

    ax.set_title(region, fontsize=11)
    ax.set_xlabel("Year")
    ax.set_xticks([2000, 2005, 2010, 2015, 2020])
    ax.grid(True, linestyle="--", alpha=0.5)

axes[0].set_ylabel("Value")
fig.suptitle("Average HAQ and Productivity Over Time by Region", fontsize=16)

fig.legend(loc="lower center", ncol=2)
plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt

# We assume df_filtered already exists

# Group to get averages per Region per Year
region_year = (
    df_filtered.groupby(["Region", "Year"])
               [["Healthcare Access and Quality Index (HAQ)",
                 "Productivity: output per hour worked"]]
               .mean()
               .reset_index()
)

regions = region_year["Region"].unique()

# Create 3 rows × 2 columns subplots
fig, axes = plt.subplots(3, 2, figsize=(14, 15))
axes = axes.flatten()

for ax, region in zip(axes, regions):
    subset = region_year[region_year["Region"] == region]

    ax.plot(subset["Year"], subset["Healthcare Access and Quality Index (HAQ)"],
            marker="o", label="HAQ")

    ax.plot(subset["Year"], subset["Productivity: output per hour worked"],
            marker="o", label="Productivity")

    ax.set_title(region, fontsize=12)
    ax.set_xlabel("Year")
    ax.set_xticks([2000, 2005, 2010, 2015, 2019])
    ax.grid(True, linestyle="--", alpha=0.5)

# Shared Y label
fig.text(0.04, 0.5, "Value", va="center", rotation="vertical", fontsize=12)

# One main title
fig.suptitle("Average HAQ and Labor Productivity by Region", fontsize=16)

# ---- SINGLE LEGEND for whole figure ----
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc="lower center", ncol=2)

# Adjust layout so legend fits
plt.tight_layout(rect=[0, 0.05, 1, 0.95])

# ---- SAVE FIGURE HERE ----
plt.savefig("HAQ_Productivity_by_Region.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# We assume df_filtered already exists

regions = df_filtered["Region"].unique()

for region in regions:

    # Subset for only this region
    reg = df_filtered[df_filtered["Region"] == region]

    countries = reg["Country"].unique()
    n = len(countries)

    # Determine subplot grid size (2 columns)
    ncols = 3
    nrows = math.ceil(n / ncols)

    fig, axes = plt.subplots(nrows, ncols, figsize=(14, 5 * nrows))
    axes = axes.flatten()  # flatten for easy iteration

    for ax, country in zip(axes, countries):
        sub = reg[reg["Country"] == country]

        # Plot HAQ
        ax.plot(sub["Year"],
                sub["Healthcare Access and Quality Index (HAQ)"],
                marker="o", label="HAQ")

        # Plot Productivity
        ax.plot(sub["Year"],
                sub["Productivity: output per hour worked"],
                marker="x", linestyle="--", label="Productivity")

        ax.set_title(country, fontsize=12)
        ax.set_xlabel("Year")
        ax.set_xticks([2000, 2005, 2010, 2015, 2019])
        ax.grid(True, linestyle="--", alpha=0.5)

    # Remove empty axes (if region has odd number of countries)
    for i in range(len(countries), len(axes)):
        fig.delaxes(axes[i])

    # Shared labels and title
    fig.text(0.04, 0.5, "Value", va="center", rotation="vertical", fontsize=12)
    fig.suptitle(f"HAQ and Productivity by Country (2000-2019)\nRegion: {region}", fontsize=16)

    # One global legend
    handles, labels = axes[0].get_legend_handles_labels()
    fig.legend(handles, labels, loc="lower center", ncol=2)

    plt.tight_layout(rect=[0, 0.05, 1, 0.95])

    # Save figure
    filename = f"Region_{region.replace(' ', '_')}_wcountries.png"
    plt.savefig(filename, dpi=300, bbox_inches="tight")

    plt.show()

    
```
